```python
import matplotlib
import seaborn
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles

figure, ax = plt.subplots(1, 1, dpi=120)
cols = seaborn.color_palette(['orange', 'darkblue', 'green'])
v_ds = venn3(subsets=(1, 1, 2, 1, 2, 2, 1), set_labels = ('', '', ''), ax=ax, set_colors=cols)

labels = [('Computer Science', 45, 'center', 'center'), 
          ('Math \& Statistics', -45, 'center', 'center'), 
          ('ML/AI', 0, 'center', 'center'), 
          ('Domain Knowledge', 0, 'center', 'bottom'), 
          ('Software', -45, 'center', 'center'), 
          ('Research', 45, 'center', 'center'), 
          ('Data Science', 0, 'center', 'center')]

def label_by_id(label, ID):
    text = v_ds.get_label_by_id(ID)
    v_ds.get_label_by_id(ID).set_text('')
    x, y = list(text.get_position())
    if ID=='100':
        x *= 0.8
    if ID=='010':
        x *= 0.8
    ax.text(x, y, label[0], rotation=label[1], ha=label[2], va=label[3])
    
for label, ID in zip(labels, ["100", "010", "110", "001", "101", "011", "111"]):
    label_by_id(label, ID)

plt.show()
```
